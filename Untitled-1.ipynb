```python
import pandas as pd
import numpy as np
from typing import Optional

def read_sensor_data(file_path: str) -> pd.DataFrame:
    """
    センサーデータファイルを読み込んでpandas DataFrameとして返す
    Parameters:
    -----------
    file_path : str
        データファイルのパス
    Returns:
    --------
    pd.DataFrame
        読み込んだセンサーデータ
        カラム: Localtime, Unixtime, AccX, AccY, AccZ, AsX, AsY, AsZ, HX, HY, HZ, AngX, AngY, AngZ
    """
    
    # データを格納するリスト
    data_rows = []
    column_names = None
    
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                line = line.strip()
                
                # 空行をスキップ
                if not line:
                    continue
                
                # ヘッダー行からカラム名を取得
                if line.startswith('##'):
                    # "## " を除去してカラム名を取得
                    column_names = line[2:].strip().split()
                    continue
                
                # コメント行をスキップ
                if line.startswith('#'):
                    continue
                
                # データ行を処理
                parts = line.split()
                
                # "No data"を含む行をスキップ
                if 'No' in parts and 'data' in parts:
                    continue
                
                # 正常なデータ行（14列必要）
                if len(parts) >= 14:
                    data_rows.append(parts[:14])  # 最初の14列のみ使用
                
    except FileNotFoundError:
        raise FileNotFoundError(f"ファイルが見つかりません: {file_path}")
    except Exception as e:
        raise Exception(f"ファイル読み込み中にエラーが発生しました: {e}")
    
    # デフォルトのカラム名（ヘッダーがない場合）
    if column_names is None:
        column_names = ['Localtime', 'Unixtime', 'AccX', 'AccY', 'AccZ', 
                       'AsX', 'AsY', 'AsZ', 'HX', 'HY', 'HZ', 'AngX', 'AngY', 'AngZ']
    
    # データが空の場合
    if not data_rows:
        return pd.DataFrame(columns=column_names)
    
    # DataFrameを作成
    df = pd.DataFrame(data_rows, columns=column_names)
    
    # データ型を適切に変換
    try:
        # Localtimeは文字列のまま
        df['Localtime'] = df['Localtime'].astype(str)
        # 'Localtime'以外のすべてのカラムを浮動小数点数に変換
        numeric_cols = [col for col in df.columns if col != 'Localtime']
        df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
        
        # Localtimeをdatetimeに変換（オプション）
        df['Localtime'] = pd.to_datetime(df['Localtime'], errors='coerce')
        
    except Exception as e:
        print(f"データ型変換中に警告: {e}")
    
    # インデックスをリセット
    df.reset_index(drop=True, inplace=True)
    
    return df
```

```python
def sensor_angle(ang_alpha, ang_beta):
    """
    センサーの角度を計算する関数
    
    Parameters:
    -----------
    ang_alpha : float
        センサーのアルファ角度
    ang_beta : float
        センサーのベータ角度
    
    Returns:
    --------
    float
        計算されたセンサーの角度
    """
    
    # ここでは単純な計算を行う例を示します
    y = np.sin(ang_alpha)
    x = np.sin(ang_beta)
    angle = np.arctan2(y, x)
    return angle
```

# arctan(sin(α)/sin(β))の誤差伝搬式

## 誤差伝搬の一般式

関数 f(α,β) = arctan(sin(α)/sin(β)) の誤差は以下の式で表されます：

$$\Delta f = \left|\frac{\partial f}{\partial \alpha}\right| \Delta \alpha + \left|\frac{\partial f}{\partial \beta}\right| \Delta \beta$$

## 偏微分の計算結果

### α に関する偏微分
$$\frac{\partial f}{\partial \alpha} = \frac{\cos(\alpha) \sin(\beta)}{\sin^2(\beta) + \sin^2(\alpha)}$$

### β に関する偏微分
$$\frac{\partial f}{\partial \beta} = \frac{-\sin(\alpha) \cos(\beta)}{\sin^2(\beta) + \sin^2(\alpha)}$$

## 最終的な誤差伝搬式

$$\boxed{\Delta f = \frac{|\cos(\alpha) \sin(\beta)|}{\sin^2(\beta) + \sin^2(\alpha)} \times \Delta \alpha + \frac{|\sin(\alpha) \cos(\beta)|}{\sin^2(\beta) + \sin^2(\alpha)} \times \Delta \beta}$$

```python
# 誤差伝搬の計算
def calculate_error_propagation(alpha, beta, d_alpha, d_beta):
    """
    arctan(sin(alpha)/sin(beta))の誤差伝搬を計算
    """
    sin_alpha = np.sin(alpha)
    cos_alpha = np.cos(alpha)
    sin_beta = np.sin(beta)
    cos_beta = np.cos(beta)
    
    # 分母: sin²(β) + sin²(α)
    denominator = sin_beta**2 + sin_alpha**2
    
    # 偏微分項
    partial_alpha = np.abs(cos_alpha * sin_beta) / denominator
    partial_beta = np.abs(sin_alpha * cos_beta) / denominator
    
    # 誤差伝搬
    error = partial_alpha * d_alpha + partial_beta * d_beta
    
    return error
```

```python
from mpl_toolkits.mplot3d import Axes3D


# 角度範囲の設定（度数）
alpha_deg = np.linspace(1, 179, 181)  # 0度と180度を避ける
beta_deg = np.linspace(1, 179, 181)
# メッシュグリッドの作成
ALPHA, BETA = np.meshgrid(alpha_deg, beta_deg)

# ラジアンに変換
alpha_rad = np.deg2rad(ALPHA)
beta_rad = np.deg2rad(BETA)

# 入力誤差（ラジアン）
delta_alpha = 0.001
delta_beta = 0.001

# 誤差の計算
ERROR = calculate_error_propagation(alpha_rad, beta_rad, delta_alpha, delta_beta)

# 度数に変換（見やすくするため）
ERROR_deg = np.rad2deg(ERROR)

# 3Dプロット
fig = plt.figure(figsize=(15, 5))

# プロット1: 誤差の3D表面
ax1 = fig.add_subplot(131, projection='3d')
surf1 = ax1.plot_surface(ALPHA, BETA, ERROR_deg, cmap='viridis', alpha=0.8)
ax1.set_xlabel(r'$\alpha$ (degree)')
ax1.set_ylabel(r'$\beta$ (degree)')
ax1.set_zlabel('error (degree)')
ax1.set_title('Error Propagation 3D Surface\n(Δα=Δβ=0.001rad)')
plt.colorbar(surf1, ax=ax1, shrink=0.5)

# プロット2: 等高線図
ax2 = fig.add_subplot(132)
contour = ax2.contourf(ALPHA, BETA, ERROR_deg, levels=20, cmap='viridis')
ax2.set_xlabel(r'$\alpha$ (degree)')
ax2.set_ylabel(r'$\beta$ (degree)')
ax2.set_title('Error Contour Map')
plt.colorbar(contour, ax=ax2)

# プロット3: 対数スケールでの3D表面
ax3 = fig.add_subplot(133, projection='3d')
ERROR_log = np.log10(ERROR_deg + 1e-10)  # 対数変換（0除算回避）
surf3 = ax3.plot_surface(ALPHA, BETA, ERROR_log, cmap='plasma', alpha=0.8)
ax3.set_xlabel(r'$\alpha$ (degree)')
ax3.set_ylabel(r'$\beta$ (degree)')
ax3.set_zlabel(r'log_{10}(error) (degree)')
ax3.set_title('Logarithmic Error Surface')
plt.colorbar(surf3, ax=ax3, shrink=0.5)

plt.tight_layout()
plt.show()
```
